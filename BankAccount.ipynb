# Bank Account Application - Object Oriented Programming

```python
from datetime import datetime
```

```python
class BankAccount:
    def __init__(self, owner):
        self.__owner = owner
        self.__balance = 0

    def get_owner(self):
        return self.__owner

    def get_balance(self):
        return self.__balance

    def __get_timestamp(self):
        now = datetime.now()
        day_name = now.strftime("%A")
        date_str = now.strftime("%d %B %Y")
        time_str = now.strftime("%I:%M%p").lower()
        return f"{day_name}, {date_str}, {time_str}"

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid amount. Must be greater than 0.")
            return
        self.__balance += amount
        timestamp = self.__get_timestamp()
        print(f"Deposited {amount} SAR to your bank account on {timestamp}.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid amount. Must be greater than 0.")
            return
        if amount > self.__balance:
            print("Insufficient balance.")
            return
        self.__balance -= amount
        timestamp = self.__get_timestamp()
        print(f"Withdrawn {amount} SAR from your bank account on {timestamp}.")

    def check_balance(self):
        print(f"{self.__owner}'s balance: {self.__balance} SAR")

    def __str__(self):
        return f"BankAccount(owner={self.__owner}, balance={self.__balance})"

    def __repr__(self):
        return self.__str__()

    # Polymorphism: + operator to transfer between accounts
    def __add__(self, other):
        total = self.__balance + other.__balance
        return total
```

```python
# --- Main ---
print("=" * 40)
print("       BANK ACCOUNT SYSTEM")
print("=" * 40)

# Create accounts
account1 = BankAccount("Khalid")
account2 = BankAccount("Sara")

# Deposit
account1.deposit(2000)
account2.deposit(5000)

print()

# Withdraw
account1.withdraw(150)

print()

# Check balance
account1.check_balance()
account2.check_balance()

print()

# Polymorphism: using + to get combined balance
combined = account1 + account2
print(f"Combined balance: {combined} SAR")

print()

# Polymorphism: using str()
print(account1)
print(account2)

print()

# Try invalid operations
account1.withdraw(999999)
account1.deposit(-100)
```
