# Bike Description - Object Oriented Programming

## class Bike

```python
class Bike:
    def __init__(self, description, cost, sale_price, condition):
        self.__description = description
        self.__cost = cost
        self.__sale_price = sale_price
        self.__condition = condition
        self.__sold = False

    # Getters
    def get_description(self):
        return self.__description

    def get_cost(self):
        return self.__cost

    def get_sale_price(self):
        return self.__sale_price

    def get_condition(self):
        return self.__condition

    def is_sold(self):
        return self.__sold

    # Methods
    def update_sale_price(self, sale_price):
        if self.__sold == True:
            print("Action not allowed, Bike has already been sold")
        else:
            self.__sale_price = sale_price

    def sell(self):
        self.__sold = True

    def display(self):
        print(f"Description: {self.__description}")
        print(f"Cost: {self.__cost}")
        print(f"Sale Price: {self.__sale_price}")
        print(f"Condition: {self.__condition}")
        print(f"Sold: {self.__sold}")
```

## Create object

```python
bike1 = Bike("Univega Alpina, orange", cost=100, sale_price=500, condition=0.5)
```

```python
print("--- Before update ---")
bike1.display()
```

```python
# Update sale price
bike1.update_sale_price(350)
print("\n--- After update ---")
bike1.display()
```

```python
 # Sell the bike
bike1.sell()
print("\n--- After selling ---")
bike1.display()
```

```python
# Try to update price after selling
print()
bike1.update_sale_price(400)
```
