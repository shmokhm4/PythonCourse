# PhoneBook Project

```python
phone_book = {
    "1111111111": "Amal",
    "2222222222": "Mohammed",
    "3333333333": "Khadijah",
    "4444444444": "Abdullah",
    "5555555555": "Rawan",
    "6666666666": "Faisal",
    "7777777777": "Layla"
}
```

## Search for a contact by phone number

```python
def search_by_number():
    phone = input("Enter the phone number: ")

    # Check if the input is valid (only digits and exactly 10 digits)
    if not phone.isdigit() or len(phone) != 10:
        print("This is invalid number")
        return

    # Check if the number exists in the phone book
    if phone in phone_book:
        print(phone_book[phone])
    else:
        print("Sorry, the number is not found")
```

## Search for a contact by name

```python
def search_by_name():
    name = input("Enter the name: ")
    found = False

    for key, value in phone_book.items():
        if value.lower() == name.lower():
            print(f"{value}: {key}")
            found = True

    if not found:
        print("Sorry, the name is not found")
```

## Add a new contact to the phone book

```python
def add_contact():
    """Add a new contact to the phone book."""
    phone = input("Enter the phone number: ")

    if not phone.isdigit() or len(phone) != 10:
        print("This is invalid number")
        return

    if phone in phone_book:
        print(f"This number already exists for '{phone_book[phone]}'")
        return

    name = input("Enter the name: ")
    phone_book[phone] = name
    print(f"Added successfully: {name} -> {phone}")
```

```python
def main():
    while True:
        print("\n===== Phone Book =====")
        print("1. Search by number")
        print("2. Search by name")
        print("3. Add new contact")
        print("4. Exit")

        choice = input("\nChoose an option: ")

        if choice == "1":
            search_by_number()
        elif choice == "2":
            search_by_name()
        elif choice == "3":
            add_contact()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid option, try again.")


if __name__ == "__main__":
    main()
```
